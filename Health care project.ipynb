```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv("C:\\Users\\NLCI\\Desktop\\Nitesh Data\\NItesh SQL Project Data\\project4\\medical_records.csv")
df=pd.DataFrame(data)
print(df)
```

```python
print("first few rows:")
print(df.head())
```

```python
print("column name:")
print(df.columns)
```

```python
print("\nsummary statistics:")
print(df.describe(include='all'))
#summary statistics
summary = df.describe()
print("mean values;\n",summary .loc['mean'])
print("\nStandard Deviation:\n",summary.loc['std'])
print("\nMinimum Values:\n",summary.loc['min'])
print("\nMaxmimum Values:\n",summary.loc['max'])
```

```python
Stroke_data = df[df['Disease_Category'] == 'Stroke']
print("Stroke_data summary:")
print(Stroke_data.describe(include='all'))
```

```python
Disease_Status =df.groupby('Disease_Category').describe(include='all')
print("Statistics by Disease_Category:")
print(Disease_Status)
```

```python
print("BOX GRAPH")
plt.figure(figsize=(12,6))
sns.boxplot(x="Disease_Category",
y="Age",data=data)
plt.xticks(rotation=45)
plt.title("Distribution of Age by Disease Category")
plt.xlabel("Disease Category")
plt.ylabel("Age")
plt.show()
```

```python
print("Violin Plot")
plt.figure(figsize=(12,6))
sns.violinplot(x="Disease_Category",
y="Age",data=data)
plt.xticks(rotation=45)
plt.title("Age Distribution by Disease Category")
plt.xlabel("Disease Category")
plt.ylabel("Age")
plt.show()
```
